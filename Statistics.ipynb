### The idea behind statistics.

#### The goal -

What is the goal of science?

* Research or scientific analysis aims to find relations between variables.
* There is no other way of representing "meaning" except in terms of relations between some quantities or qualities.
* Typically this involves measuring from a sample of observations.

What is a relation?

* In a sample of observations, we find a relation if values are distributed in a consistent manner.
* Variables are related if their values systematically correspond to each other for these observations.

How can we find relations?

(1) Correlational research: 
    
* We do not influence any variables (what we measure) by design. 
* We measure variables and look for relations (correlations) between them.
* Cannot conclusively prove causality.
 
(2) Experimental research:
    
* Manipulate a specific set of variables.
* Measure the effects of this manipulation on other variables.

What is statistics?

* A set of tools for evaluating relations between variables.
* Involves taking measurments from a sample.
* Using the sample to infer something about the population.

#### Types of goals - 

(1) Intervention (group differences) - T-test, anova, chi square, Wilcoxon.
  
(2) Description (association) - Correlations, factor analysis, path analysis, unsupervised learning.

(3) Explanation (prediction) - Regression, logistic regression, discriminant analysis.

### Intervention.

#### Sampling and the law of large numbers - 

The challenge of sample:

* We want to know a characteristic of a population, a parameter.
* We take a sample from the population and compute a quantity called a statistic, which differs by sampling error.
* If we re-sample, will we observe the same relation? 

Of course, as the sample size gets larger the sample statistics will more closely match the population parameters.

This is because of the law of large numbers.

$ \frac{1}{N} \sum\limits_{i=1}^N Z_{i} \rightarrow E[Z], N \rightarrow \infty $

To see this, let's sample 100000 Poisson random variables with parameter $\lambda$=4.5. 

The expected value of a Poisson random variable is equal to it's parameter.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd
from scipy import stats
%matplotlib inline

# Create a random sample 
sample_size = 100000
lambda_ = 4.5
N_samples = range(1, sample_size, 100) # 1000 samples
colors = matplotlib.rcParams['axes.color_cycle']
for k,c in zip(range(3),colors):
    samples = np.random.poisson(lam=lambda_, size=sample_size) # Draw samples from a Poisson distribution.
    partial_average = [samples[:i].mean() for i in N_samples] # Draw increacingly large N 
    plt.plot(N_samples, partial_average, lw=2, color=c, label="Average of  $n$ samples; Trial: %d" % k)
plt.plot(N_samples, lambda_ * np.ones_like(partial_average),ls="--", label="Expected value", c="k")
plt.ylim(4.35, 4.65)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title("Convergence of the average of random variables to its expected value",fontsize=8)
plt.ylabel("Sample mean for $n$ samples",fontsize=8)
plt.xlabel("# of samples, $n$",fontsize=8)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':8})
plt.grid()
```

#### Confidence intervals - 

So, we have a population. We sample from it. We estimate a parameter from the sample. 

Confidence intervals tell us how well we determined the population parameter!

(Note that this is different from prediction intervals, which tell us something about the next sample.)

For example, lets say we are using sample mean as an estimator of a population mean.

* We have a sample statistic ($u_s$) for sample size $n$ and population size $N$.
* Knowing the standard deviation of our sample, $s$, we can work out the Standard Error of our sampling statistic.

For estimating means, this is simply the standard error of the mean (SEM).

When $n$ << $N$, then the Standard Error of our sampling statistic is $SE = \frac{s}{\sqrt{n}}$.

Now, a key point: the distribution of this sampling statistic is often normal:

* If course, this is true if the population is normal.
* But, it's also true if the sample size is sufficiently large.

The latter points is known from the central limit theorm!

Given a population with $any$ distrbution, take samples of size $s$ from it.

Plot the frequency of sample means. 

As $s$ increases:

(1) The histogram of sample means will become more normally distributed with $u_s = \mu_{population}$.

(2) This means that the mean of the sampling distribution will be equal to the mean of the population distribution. 

(3) The standard deviation gets smaller. 

(4) Knowing $s$ and $\sigma_{population}$, we can predict the standard deviation of the sampling distribution of the sample mean:

$$\frac{\sigma_{population}}{\sqrt{N}}$$

```python
ndata = 1e5 # 100k random datapoints between 0 and 1
lambda_ = 4.5 # expected value of 4.5
data = np.random.poisson(lam=lambda_, size=ndata) # draw Poisson random variables
colors = matplotlib.rcParams['axes.color_cycle']
sampleSize=[10,50,100] 
for ss,c in zip(sampleSize,colors):
    sumData=np.mean(data.reshape(int(ndata/ss),ss),axis=1) # sample means
    hist,bins=np.histogram(sumData, bins=np.linspace(0,np.max(sumData),50))
    center=(bins[:-1] + bins[1:])/2
    plt.plot(center,np.array(hist),linewidth=2,color=c,label='Sample size %s'%ss)
plt.xlim([2,7])
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('Sample mean',fontsize=8)
plt.ylabel('Counts per bin',fontsize=8)
plt.title('Sampling distribution of sample mean',fontsize=8)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':8})
```

So, in many cases we know that the sample statistic is about normally distributed!

No, it's not quite. 

That's why a different distribution - the t-Distrbution - is used. 

This looks closer to normal as $N$ gets bigger.

Either way, we can compute the density around sample mean for our t-Distribution.

For normally distributed variables:

* Recall the $68–95–99.7$ rule.
* Recall z-score indicate the number of standard deviations from the mean $ Z=\frac{(x-\mu)}{\sigma} $.

So, $95\%$ of the data will be $z=2$ from the mean in both directions.

Then, we can compute the $95\%$ confidence interval bounds using and the computed $SE$:

* T is just the specified percentile of the Student's t-distribution.
* This is analogous to $95%$ of the data being $1.96 \times \sigma$ for a normal distribution!  
* Lower bound is simply $u_s - T \times SE$.

Simply ways to understand this:

* If we run the expariment 100 times, 95 of the sample means will be between these bounds.
* There is 95% probability that the true mean (the mean of the population) is inside this interval.

Now, it's also critical to see that these bounds get smaller $n$ gets bigger!

Note that confidence intervals in practice are computed using this principles along with re-sampling methods:

http://en.wikipedia.org/wiki/Bootstrapping_%28statistics%29#Deriving_confidence_intervals_from_the_bootstrap_distribution

Finally, we can easily compute this for a sample with ($\mu=0$,$\sigma^2=1$) with size equals to 10.

```python
stats.t.interval(alpha=0.95, df=10-1, loc=0, scale=1) / np.sqrt(10)
```

In short, what's the point?

(1) We use samples to estimate population parameters.

(2) Confidence intervals are a reasonable framework for evaluating how well the statistic estimates the population parameter.

(3) The confidence bounds depend largely on the size of our sample.

#### Error modes -

So, we know that that sample estimates for parameters are a funtion of the sample size.

This could be affected in two ways:

(1) We observe a relation that occoured by chance -

* We address this with statistical signifiance.
* We use p-values to evaluate this, which is the likelihood that the observed relation occoured by chance.
* Thresholds are used to evaluate p-values.

(2) We miss a bona fide relation -

* We address this with power analysis.


### Statistical signifiance.

#### Did a relation occour by chance?

In the population, there is some true relation between variables.

Two things matter here:

* Sample size.
* Effective size.

If the relation is small, then there is no way to identify it unless the sample size is correspondingly large!

If the relation is large, then it can be found to be significant even in a study based on a very small sample.

When the magnitude of the effect approaches 0, the necessary sample size to conclusively prove it approaches infinity.

#### Null-hypothesis testing -

So, we want to detect a differece between groups. 

Let's say we want to compare means.

Of course, we could compute the differnce statistic with confidence intervals, as discussed above.

This notebook has an illustration of that (and there's likely a funtion to do this in scipy):
http://nbviewer.ipython.org/github/aok1425/stats-tests/blob/master/frequentist.ipynb

But, another approach is more binary: we run a test. 

Null hypothesis significance testing (NHST) is common:

* State a null hypothesis ($H_o$): There is no effect.
* Calculate the $p$ value: The probability of seeing getting our result given $H_o$.
* If $p$ is sufficiently small, we reject the null hypothesis.


#### Parametric tests: t-Test - 

If we know something about our distributions, we can use a parametric test.

This is a method to evaluate the differences in means between two groups.

Assumptions:
    
* Normally distributed sample statistic! 
* That is why the test is parametric.
* Two groups.

The Central Limit Theorm tells us that the sample statistic is normal if the sample size is sufficiently large.

This can work on small sample sizes if:

* Variables are normally distributed within each group.
* The variation between groups is not reliably different (e.g., examine with the F test).
    
The test will output a probability of error associated with rejecting the hypothesis of no difference categories.

So, let's fire it up with random Gaussian data and see.

```python
import scipy
import seaborn as sns
s_1 = np.random.normal(1, 1, 100)
s_2 = np.random.normal(4, 1, 100)
dat=pd.DataFrame(np.array([s_1,s_2]).T)
sns.violinplot(dat)
scipy.stats.ttest_ind(dat[0],dat[1], axis=0, equal_var=True)
```

```python
s_1 = np.random.normal(1, 1, 100)
s_2 = np.random.normal(1, 1, 100)
dat=pd.DataFrame(np.array([s_1,s_2]).T)
sns.violinplot(dat)
print scipy.stats.ttest_ind(dat[0],dat[1], axis=0, equal_var=True)
```

OK, that works.

Alternatives:

* If making multi-group comparisons, use ANOVA. For this, we can simply bin groups by a categorical variable.
* If normality is not certain, use non-parametric tests (below).

#### Problem of normality -

Assume we are trying to estimate the population mean.

To use the T-test (and many other parametric tests), our sample statistic needs to be normally distributed. 

This can happen in two ways:

* We know that the population is normally distributed.
* Our sample size is sufficiently large (e.g., $>100$), even if the population is not Gaussian, due to Central Limit Theorm.

That said, the T-test is resiliant to normality constrain. 

http://nbviewer.ipython.org/github/dolaameng/tutorials/blob/master/ipython-notebooks/statistics.ipynb

But, I usually use non-parametric test (as I've typically had a lot of data!).

#### Non-parametric tests: Wilcoxon, Chi-Squared -

We can't use tests that rely on a normally distributed statistic if we have small sample sizes and aren't sure the population is normal!

In short, we need statistical procedures that enable us to process:

* From small samples.
* On variables about which nothing is known (concerning their distribution).

Again, the goal is the same as above.

We want to compare means between two groups, but we know nothing about the distrbutions.

For continuous data, we can use (in place of T-test):

* Wald-Wolfowitz runs test
* The Mann-Whitney U test
* The Kolmogorov-Smirnov two-sample test

For categorical data we can use Chi-Squared test.

```python
s_1 = np.random.normal(1, 1, 100)
s_2 = np.random.normal(4, 1, 100)
dat=pd.DataFrame(np.array([s_1,s_2]).T)
sns.violinplot(dat)
scipy.stats.ranksums(dat[0],dat[1])
```

### Statistical power.

#### Detailed view of exparimental design - 

There are basically two general issues to which Experimental Design is addressed:

(1) How to design an optimal experiment: 

* Allow the experimenter to evaluate the consequences of changing the settings of a particular factor.

(2) How to analyze the results of an experiment.

There is a detailed literature on exparimental design borne out of process manufacturing:
    
http://www.statsoft.com/Textbook/Experimental-Design

#### Simple view of exparimental design - 

Statistical power analysis and sample size estimation:

* Allow you to decide how large a sample is needed to enable statistical judgments that are accurate and reliable.
* How likely your statistical test will be to detect effects of a given size in a particular situation.

Without these calculations, sample size may be too high (costly) or too low (inaccurate).

#### Error - 

Our statistic will not be equal to the actual parameter due to sampling error!

We know that sampling error is a funtion of $N$, the sample size.

The larger the sample size $N$, the smaller the error.

#### Testing - 

We have a hypothesis.

We can measure a statistic in a sample to test this.

Reject-Support testing: rejecting the null hypothesis supports the experimenter's theory.

Key parameters:

* Type I Error Rate ($\alpha$) - The probability of incorrectly rejecting a true statistical null hypothesis (false positive).
* Type II Error Rate ($\beta$) - The probability of incorrectly accepting a false statistical null hypothesis (false negative).
* Power - $1-\beta$

If you have more power if there's a lower likelihood of missing a true relation!

Also, recall that $P$ is the probability that the null hypothesis is true (no relation).

We use $0.05$ as a common cut-off.

Note that we often apply the samme cutoff of $0.05$ to $\alpha$.

We use this cutoff because, from society's standpoint, false positives are particularly undesirable.

#### Power - 

Properly designed experiments must ensure that power is high enough to detect reasonable departures from the null hypothesis.

A few key factors must be considered:

* What kind of statistical test is being performed. 
* Sample size. In general, the larger the sample size, the larger the power. 
* The size of experimental effects. 
* The level of error in experimental measurements (e.g., "noise").

We have a population parameters $\pi$ and test statistic $s$.

We want to know whether $\pi>0.5$. So, or null hypothesis is that $\pi \leq 0.5$.

The typical way to do this is a statistical test that computes a $p-value$ associated with our test statistic $s$.

The other way to look at it is that we define a threshold at which this is rejected.

For example, $s=0.58$ is the threshold at which $\alpha$ is $< 0.05$ assuming $\pi \leq 0.5$.

Recall that $s$ is computed from a sample of size $N$.

The distribution of $s$:

* Is normal for large $N$ given by central limit theorm 
* As $N$ is larger the standard error of the sample statistic gets tighter!

So, here's the problem: let's say $\pi = 0.55$. We should reject the nully hypothesis.

But, there's a distribution of $s$ values given by $N$! We don't always get an $s$ that exceeds our cutoff (above)!

So, we simply state that there's inadequate power to detect a departure of 5 percentage points from the null hypothesized value.

This all boils down to the width of the distributions:

* There are two: one under the null ($\pi=0.5$) for the actual measured test statistic.
* Both get tighter around the actual value of $\pi$ as $N$ gets bigger!
* So, more subtle differences between them are possible to detect.

http://www.statsoft.com/Textbook/Power-Analysis

### Alternative approaches.

Challenges ---

(1) Sampling is sensitive to $N$

* Repeat your experiment and you’ll get a p value that could be extremely different.

(2) Tests are binary outcomes.

* It presents the world as black or white — an effect is statistically significant or not.

(3) Multiple testing.

* We might randomly find a "significant" association if we run many expariments.

---

Rather than testing, we could:

(1) Just use estimation and report data with confidence intervals, rather than running a statistical test.

* An estimated change between exparimental groups with a confidence interval.
* A confidence interval gives us the best estimate of the true effect.

(2) Bayesian thinking ---

* Frequentist thinking: above we compute statistics from samples, which are approximations of a larger (true) population.
* With Bayes, we can make a direct probability statement.
* This is more intutive than p-values or confidence intervals.
