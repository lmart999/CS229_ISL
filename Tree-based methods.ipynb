#### The case for trees -

Consider that the relationship between the features and the response is well approximated by a linear model:

* Then, an approach such as linear regression will likely work well and will outperform a method such as a regression tree.

If instead there is a highly non-linear and complex relationship between the features:

* Then, decision trees may outperform classical approaches (bottom).

```python
import os
from IPython.display import Image
i = Image(filename=os.getcwd()+'/Images/Trees.png') # from the ISL text.
i
```

In short, trees provide:
    
* A set of rules for splitting the feature space into a set of regions, $R_1$ to $R_j$.
* For every observation that falls into the region $R_j$ , we make the same prediction,

This means that we divide the predictor space into high-dimensional rectangles, or boxes.

We can use:

* Regression tree: for numerical data.
* Classification tree: for categorical data.

Nice things about trees:

* Trees are very easy to explain to people.
* Trees can be displayed graphically, and are easily interpreted even by a non-expert.
* Trees can easily handle qualitative predictors without the need to create dummy variables.


#### Problem with trees and case for bagging -

Trees suffer from suffer from high variance:
    
* They are likely to over-fit training data.

Methods for reducing the variance of a statistical learning method is bootstrap aggregation or Bagging.

The key idea: averaging a set of observations reduces variance, $ \frac{\sigma^2}{n}$.
    
A natural way to reduce the variance and hence increase the prediction accuracy of a statistical learning method: 

* Take many training sets from the population.
* Build a separate prediction model using each training set.
* Average the resulting predictions. 

In short, we compute our parameters using a set of different bootstrapped training sets and average across them.

This is particularly useful for decision trees. 


#### Random forrests - 

Variance is a problem with trees.

We reduce this by bagging, meaning that we simply the training data up, train on sub-set, and average the resulting rules.

But, what is there is one strong predictor on the data.

Then, all or most of the bagged trees will use this strong predictor in the top split. 

Consequently, all of the bagged trees will look quite similar to each other. 

Hence the predictions from the bagged trees will be highly correlated. 

Averaging many highly correlated quantities does not lead to as large of a reduction in variance.

Rather, we want to perform averaging over many uncorrelated quantities. 

Random forests overcome this problem by forcing each split to consider only a subset of the predictors. 

It uses a small tweak that decorrelates the trees:
    
* As in bagging, we build a number of decision trees on bootstrapped training samples. 
* But at each split in the tree, the algorithm is not even allowed to consider a majority of the available predictors. 

Given a full set of $p$ predictors, a sample of $m$ predictors is taken at each split.

On average $(p − m)/p$ of the splits will not even consider the strong predictor.

So, the yet parameter is $m$: if $m = p$, then this amounts simply to bagging.

#### Parameter estimation - 

One can obtain an overall summary of the importance of each predictor:
    
* Using the RSS (for bagging regression trees) 
* The Gini index (for bagging classification trees)

#### Implementation for trees - 

```python
import sys,os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# 2D Gaussian with mean and covariance
X_sig= np.random.multivariate_normal(mean=[0.0, 0.0], cov=[ [1,0.5], [0.5,1] ], size= 1000)
print X_sig.shape
```

Here is our signal.

```python
sns.jointplot(X_sig[:,0],X_sig[:,1], kind="hex")
```

Here is our background.

```python
# Random numbers between 0 and 1 and fill them in an array of a given shape
X_bkg = np.random.rand(1000,2)
X_bkg = (X_bkg-0.5) * 10 # Re-center and re-scale
print X_bkg.shape
```

```python
sns.jointplot(X_bkg[:,0],X_bkg[:,1], kind="hex")
```

Let's merge them.

```python
# Merge signal and background samples
X_train = np.concatenate([X_bkg, X_sig])
print X_train.shape

# Create an array to hold their class indices
c_train = np.array( [0]*len(X_bkg) + [1]*len(X_sig) )
print c_train.shape
```

```python
dat=pd.DataFrame(X_train)
dat.columns=['X1','X2']
dat['Class']=c_train
dat.head(3)
```

```python
g = sns.FacetGrid(dat, hue="Class", palette="Set1", size=5)
g.map(plt.scatter, "X1", "X2", edgecolor="white", s=50, lw=1)
```

Nice. Clearly a weird decision boundary. 

A linear model would not work well.

A tree-based method might, through.

Let's use a `DecisionTreeClassifier`:

http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier

```python
# Create tree object to train
from sklearn import tree
clf=tree.DecisionTreeClassifier(min_samples_split= 20, min_samples_leaf= 10)

# Train / fit to the training sample
clf.fit(X_train, c_train)
scores = clf.score(X_train, c_train)
print "The accuracy is: %s"%scores

# Create a meshgrid as our test data
plot_step= 0.05
xmin, xmax= X_train[:,0].min(), X_train[:,0].max()
ymin, ymax= X_train[:,1].min(), X_train[:,1].max()
xx, yy = np.meshgrid(np.arange(xmin, xmax, plot_step), np.arange(ymin, ymax, plot_step) )

# Re-cast every coordinate in the meshgrid as a 2D point
X_test= np.c_[xx.ravel(), yy.ravel()]

# Predict the class
Z = clf.predict( X_test )

# Re-shape the results
Z= Z.reshape( xx.shape )
cs = plt.contourf(xx, yy, Z, cmap= plt.cm.winter, alpha=0.3)

# Overlay training samples
plt.scatter(X_bkg[:,0], X_bkg[:,1], c='b', alpha=0.5)
plt.scatter(X_sig[:,0], X_sig[:,1], c='g', alpha=0.5)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('Feature 1',fontsize=5)
plt.ylabel('Feature 2',fontsize=5)
plt.title("Boundary for decision tree classifier",fontsize=7.5)
```

#### Let's try regression trees - 

The idea is similar to classification.

However, the model prediction is given by the mean response value for the training examples in the split bin.

Note that we use a `DecisionTreeRegressor`:

http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html#sklearn.tree.DecisionTreeRegressor

```python
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

We have a non-linear relationship between the features and target!

```python
plt.scatter(X,y)
```

Let's use a `regression tree`.

Note that we tune the depth of the three using `max_depth`.

```python
from sklearn.tree import DecisionTreeRegressor
clf_2 = DecisionTreeRegressor(max_depth=5)
clf_2.fit(X, y)
scores = clf_2.score(X, y)
print "The r^2 is: %s"%scores
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_2 = clf_2.predict(X_test)
plt.figure()
plt.scatter(X, y, c="k", label="data")
plt.plot(X_test, y_2, c="r", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
```

Importances are useful if we have many features.

```python
clf_2.feature_importances_
```

#### Implementation for random forrest - 

```python
# Load data
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
pair=[0, 1]
X = iris.data[:, pair]
y = iris.target
```

```python
idx = np.arange(X.shape[0])
np.random.shuffle(idx)
X = X[idx]
y = y[idx]
```

```python
mean = X.mean(axis=0)
std = X.std(axis=0)
X = (X - mean) / std
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import clone
n_estimators = 30
model = RandomForestClassifier(n_estimators=n_estimators)
clf = model.fit(X, y)
scores = clf.score(X, y)
```

The classification accuracy. 

```python
print scores
```

Let's get the importance of the features. The higher, the more important the feature:
    
* The importance of a feature is computed as the (normalized) total reduction of the criterion brought by that feature. 
* It is also known as the Gini importance 

```python
print clf.feature_importances_
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))
```

```python
cmap = plt.cm.RdYlBu
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=cmap)
```

```python
plot_colors = "ryb"
n_classes = 3
for i, c in zip(xrange(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=c, label=iris.target_names[i],cmap=cmap)
```

```python
cs = plt.contourf(xx, yy, Z, cmap=cmap)
for i, c in zip(xrange(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=c, label=iris.target_names[i],cmap=cmap)
```
